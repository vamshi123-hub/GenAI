```python
#Prompt Template
from langchain.chains import LLMChain
from langchain import PromptTemplate

generic_template = """
    Write a Summary of following speech:
    Speech:{speech}
    Translate the precise summary to {language}
"""
```

```python
prompt = PromptTemplate(
    input_variables = ["speech", "language"],
    template = generic_template
)
prompt
```

```python
complete_prompt = prompt.format(speech = speech, language = "Telugu")
complete_prompt
```

```python
llm.get_num_tokens(complete_prompt)
```

```python
llm_chain = LLMChain(llm = llm, prompt = prompt)
summary = llm_chain.run({"speech" : speech, "language" : "Telugu"})
summary
```

```python
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader("apjspeech.pdf")
docs = loader.load_and_split()
docs
```

```python
template = """
    Write a Concise and Short Summary of Following Speech,
    Speech:{text}
"""
prompt = PromptTemplate(
    input_variables = ["text"],
    template = template
)
```

```python
from langchain.chains.summarize import load_summarize_chain
chain = load_summarize_chain(
    llm = llm,
    chain_type = "stuff",
    prompt = prompt,
)
output_summary = chain.run(docs)
output_summary
```

```python
#Map Reduce to Summarize the Large Documents
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 1500, chunk_overlap = 200)
final_document = text_splitter.split_documents(docs)
final_document
```

```python
len(final_document)
```

```python
chunk_prompt = """
    Please Summarize the below speech:
    Speech:{text}
    summary :
"""

map_prompt_template = PromptTemplate(
    input_variables = ["text"],
    template = chunk_prompt
)
```

```python
final_prompt = """
    Provide the final summary of the entire speech with these important points.
    Add a Motivation Title, Start the precise Summary with an Introduction and Provide the Summary in number points for the speech.
    Speech:{text}

"""
final_prompt_template = PromptTemplate(
    input_variables = ["text"],
    template = final_prompt
)
final_prompt_template
```

```python
summary_chain = load_summarize_chain(
    llm,
    chain_type = "map_reduce",
    map_prompt = map_prompt_template,
    combine_prompt = final_prompt_template,
    verbose = True,
)
final_output = summary_chain.run(final_document)
final_output
```

```python
#Refine Chain Summary
chain = load_summarize_chain(
    llm,
    chain_type = "refine",
    verbose = True,
)
refine_summary = chain.run(final_document)
refine_summary
```
